## 栈与队列

栈与队列都是用于处理<span style="color: yellow">临时数据</span>的重要数据结构

### 栈

- 只能在末尾插入数据
- 只能读取末尾的数据
- 只能移除末尾的数据

### 栈的作用

```python
str_to_be_checked: str = "[x for x in range(10])"
```

```python
class StackCheckBlanket:
    def __init__(self, string: str) -> None:
        """
        :param string: 被检测字符串的初始化
        """

        self.string: str = string

    def check(self) -> bool:
        """
        检查括号是否匹配，返回检查结果

        :return: 检测结果
        """

        check_stack: list[str] = []
        for char in self.string:
            # 如果是左半边括号，就入栈
            if char in ["(", "[", "{"]:
                check_stack.append(char)
            # 如果是右半边括号，就出栈，并检查是否正确
            elif char in [")", "]", "}"]:
                # 如果栈是空的，说明不匹配，返回False
                if not check_stack:
                    return False
                match char:
                    case ")":
                        if check_stack[-1] != "(":
                            return False
                        else:
                            check_stack.pop()
                    case "]":
                        if check_stack[-1] != "[":
                            return False
                        else:
                            check_stack.pop()
                    case "}":
                        if check_stack[-1] != "{":
                            return False
                        else:
                            check_stack.pop()
        return not check_stack
```

```python
stack_checker = StackCheckBlanket(str_to_be_checked)
stack_checker.check()
```

```python
maze = [
      "#########",
      "#S......#",
      "#.#####.#",
      "#.....#.#",
      "#####.###",
      "#......E#",
      "#########"
  ]
```

```python
# 方向：上、下、左、右
dirs = [(-1,0), (1,0), (0,-1), (0,1)]

def find_path(maze):
    stack = [(1, 1)]          # 栈：存路径，起点入栈
    visited = set([(1, 1)])   # 记录已走过的地方

    while stack:
        x, y = stack[-1]      # 看栈顶（当前位置）

        # 找到终点！
        if maze[x][y] == 'E':
            return stack      # 栈里存的就是完整路径

        # 找下一步
        found = False
        for dx, dy in dirs:
            nx, ny = x + dx, y + dy

            # 能走（是通路且没走过）
            if maze[nx][ny] != '#' and (nx, ny) not in visited:
                stack.append((nx, ny))     # 入栈 = 向前走
                visited.add((nx, ny))
                found = True
                break                      # 找到一个方向就继续

        if not found:
            stack.pop()                    # 出栈 = 回溯！

    return None  # 无路可走
```

```python
find_path(maze)
```

### 队列

- 只能在末尾插入数据(与栈相同)
- 仅在开头读取数据(与栈不同)
- 仅在开头移除数据(与栈不同)

<img src="./pics/stack-1.png" width="700" />

仍然是这样一个数组

<img src="./pics/stack-2.png" width="700" />

向队列内添加数据的操作与入栈相同

从队列中读取和移除数据的操作与出栈不同

<img src="./pics/queue-3.png" width="700" />

```python
class MyQueue:
    def __init__(self) -> None:
        """
        初始化队列
        """

        self.queue: list[Any] = []

    def enqueue(self, item: Any) -> None:
        """
        向队列中添加数据

        :param item: 添加的数据
        """

        self.queue.append(item)

    def dequeue(self) -> Any:
        """
        从队列中移除数据并返回

        :return: 移除的数据
        """

        return self.queue.pop(0)
```

```python
queue = MyQueue()
```

```python
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
```

```python
queue.dequeue()
```

## 双端队列

- 可以在头和尾插入和删除数据
- 同侧就是栈
- 两侧就是队列

```python
from collections import deque

dq = deque([1,2,3,4])
```

```python
dq.append(1)

dq
```

```python
dq.pop()
```

```python
dq
```

```python
dq.popleft()
```

```python
dq
```

```python
dq.appendleft(5)

dq
```
