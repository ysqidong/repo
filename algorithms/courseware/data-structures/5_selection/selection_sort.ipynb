## 选择排序

### 规则

- 将数组分为两个部分：已排序部分（初始为空）和未排序部分（初始为整个数组）。
- 在未排序部分中，找到最小（或最大，取决于升序/降序）元素。
- 将这个最小元素与未排序部分的第一个元素交换位置。
- 将已排序部分的边界向后移动一位。
- 重复步骤2-4，直到未排序部分只剩一个元素（或为空），排序完成。

```python
class SelectionSort:
    def __init__(self, arr: list[int|float]):
        """
        :param arr: 初始化数组
        """
        self.arr = arr
        self.len = len(self.arr)

    def sort(self) -> list[int|float]:
        """
        选择排序实现

        :return: 返回排序后数组
        """

        for i in range(self.len - 1):
            min_index = i
            min_value = self.arr[i]

            # 循环内找到最小元素
            for j in range(i + 1, self.len):
                if self.arr[j] < min_value:
                    min_index = j
                    min_value = self.arr[j]

            # 移动最小元素的位置
            self.arr[i], self.arr[min_index] = self.arr[min_index], self.arr[i]

        return self.arr
```

```python
selection_sort = SelectionSort([6, 5, 7, 2, 1])
selection_sort.sort()
```

### 冒泡排序与选择排序的实际执行步数上有差别吗？

假设数组 [1, 2, 3, 4, 5]<span style="color: yellow">（已有序）</span>：

- 选择排序：10 次比较，0 次交换
- 冒泡排序：4 次比较，0 次交换

假设数组 [5, 4, 3, 2, 1]<span style="color: yellow">（逆序）</span>：

- 选择排序：10 次比较，4 次交换
- 冒泡排序：10 次比较，10 次交换


随机数据：都是$O(N^2)$，但是选择排序交换更少（冒泡排序只能相邻交换）
