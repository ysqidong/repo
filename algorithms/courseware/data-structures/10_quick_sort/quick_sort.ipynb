# 快速排序 (Quick Sort)

### 回顾：我们学过的排序算法

| 算法 | 最好情况 | 平均情况 | 最坏情况 |
|------|----------|----------|----------|
| 冒泡排序 | O(N) | O(N²) | O(N²) |
| 选择排序 | O(N²) | O(N²) | O(N²) |
| 插入排序 | O(N) | O(N²) | O(N²) |

有没有更快的排序算法呢？

## 分而治之的思想

快速排序采用<span style="color: yellow">分而治之（Divide and Conquer）</span>的策略：

1. **分（Divide）**：将大问题分解成小问题
2. **治（Conquer）**：递归地解决小问题
3. **合（Combine）**：将小问题的解合并成大问题的解

快速排序的核心是<span style="color: yellow">分区（Partition）</span>操作。

### 分而治之的过程

- 原始问题：大数组
- 分解：分区成左右子数组
- 解决：递归排序子数组
- 合并：无需合并，原地有序

## 什么是分区？

分区指的是：<span style="color: yellow">选择一个轴（Pivot），将数组分为两部分，左边都比轴小，右边都比轴大。</span>

假设有一个数组：`[0, 5, 2, 1, 6, 3]`

我们选择最后一个元素 **3** 作为轴：

```
原始: [0, 5, 2, 1, 6, 3]
                   轴 ↑

分区后: [0, 2, 1]  [3]  [5, 6]
        小于3      轴    大于3
```

注意：分区完成后，<span style="color: yellow">轴的位置已经确定</span>（3已经在正确位置），但左右两边的元素顺序还未确定。

## 分区的详细步骤

以数组 `[0, 5, 2, 1, 6, 3]` 为例，轴为最后一个元素 **3**：

**步骤1**：设置左右指针
- 左指针从数组开头（索引0）开始
- 右指针从轴的左边（倒数第二个）开始

```
[0, 5, 2, 1, 6, 3]
 ↑           ↑  ↑
左(0)      右(6) 轴(3)
```

**步骤2**：左指针向右移动，直到找到大于或等于轴的值
- 0 < 3，继续右移 → 停在 **5**（5 >= 3）

**步骤3**：右指针向左移动，直到找到小于或等于轴的值
- 6 > 3，左移 → 停在 **1**（1 <= 3）

**步骤4**：交换左右指针指向的值
- 交换 5 和 1 → `[0, 1, 2, 5, 6, 3]`

**步骤5**：重复步骤2-4，直到左右指针相遇或交叉
- 左指针继续右移，停在 **5**（5 >= 3）
- 右指针继续左移，停在 **2**（2 <= 3）
- 此时左指针在右指针右边，停止

**步骤6**：将左指针的值与轴交换
- 交换 5 和 3 → `[0, 1, 2, 3, 6, 5]`

分区完成！3已经在正确位置。

````python
from collections.abc import Sequence


class QuickSort:
    """
    快速排序类，包含分区和排序功能。
    """

    def __init__(self, arr: Sequence[int | float]):
        """
        初始化快速排序器。

        :param arr: 待排序的数组，可以是整数或浮点数
        """
        self.arr: list[int | float] = list(arr)
        self.length: int = len(arr)

    def partition(self, left_index: int, right_index: int) -> int:
        """
        对数组的指定范围进行分区操作。

        选择一个轴（这里选择最右边的元素），将数组分为两部分：
        - 左边的元素都小于轴
        - 右边的元素都大于轴

        :param left_index: 分区范围的左边界索引
        :param right_index: 分区范围的右边界索引
        :return: 轴的最终位置索引
        """
        # 选择最右边的元素作为轴
        pivot_index: int = right_index
        pivot: int | float = self.arr[pivot_index]

        # 右指针从轴的左边开始
        right_pointer: int = right_index - 1
        left_pointer: int = left_index

        while True:
            # 左指针向右移动，直到找到大于或等于轴的值
            while left_pointer <= right_pointer and self.arr[left_pointer] < pivot:
                left_pointer += 1

            # 右指针向左移动，直到找到小于或等于轴的值
            while left_pointer <= right_pointer and self.arr[right_pointer] > pivot:
                right_pointer -= 1

            # 如果左右指针相遇或交叉，停止
            if left_pointer >= right_pointer:
                break

            # 交换左右指针指向的值
            self.arr[left_pointer], self.arr[right_pointer] = (
                self.arr[right_pointer],
                self.arr[left_pointer],
            )

        # 最后将左指针的值与轴交换
        self.arr[left_pointer], self.arr[pivot_index] = (
            self.arr[pivot_index],
            self.arr[left_pointer],
        )

        # 返回轴的最终位置
        return left_pointer

    def sort(
        self, left_index: int | None = None, right_index: int | None = None
    ) -> list[int | float]:
        """
        对数组进行快速排序。

        算法步骤：
        1. 对数组进行分区，将轴放到正确位置
        2. 对轴左边的子数组递归排序
        3. 对轴右边的子数组递归排序

        :param left_index: 排序范围的左边界索引，默认为数组开头
        :param right_index: 排序范围的右边界索引，默认为数组结尾
        :return: 排序后的数组
        """
        # 初始化边界
        if left_index is None:
            left_index = 0
        if right_index is None:
            right_index = self.length - 1

        # 基准情形：如果子数组长度为0或1，不需要排序
        if right_index - left_index <= 0:
            return self.arr

        """
        长度为0： sort(0, -1)
        长度为1： sort(0, 0)
        长度为2： sort(0, 1)
        """

        # 对当前范围进行分区，获取轴的位置
        pivot_position: int = self.partition(left_index, right_index)

        # 对轴左边的子数组递归排序
        self.sort(left_index, pivot_position - 1)

        # 对轴右边的子数组递归排序
        self.sort(pivot_position + 1, right_index)

        return self.arr
````

````python
# 测试分区功能
test_arr: list[int] = [0, 5, 2, 1, 6, 3]
print(f"原始数组: {test_arr}")

qs: QuickSort = QuickSort(test_arr)
pivot_pos: int = qs.partition(0, len(test_arr) - 1)

print(f"分区后数组: {qs.arr}")
print(f"轴的位置: {pivot_pos}")
print(f"轴的值: {qs.arr[pivot_pos]}")
````

## 快速排序的递归过程

>

1. 第一次分区：整个数组分区，轴3到位
2. 左子数组：[0,2,1] 继续分区
3. 右子数组：[6,5] 继续分区
4. <span style="color: yellow">递归直到</span>子数组长度为0或1

````python
# 测试完整快速排序
test_arr2: list[int | float] = [64, 34, 25, 12, 22, 11, 90, 5.5, 3.14]
print(f"排序前: {test_arr2}")

qs2: QuickSort = QuickSort(test_arr2)
sorted_arr: list[int | float] = qs2.sort()

print(f"排序后: {sorted_arr}")
````

## 快速排序的时间复杂度分析

### 分区操作的复杂度

每次分区操作：
- 比较：每个元素都要与轴比较一次 → **N次比较**
- 交换：最少1次（<span style="color: yellow">顺序</span>），最多N/2次（<span style="color: yellow">逆序</span>），平均约N/4次
- 总计：约 **1.25N** 步 → 根据大O规则，为 **O(N)**

### 完整快速排序的复杂度

**平均情况（每次轴都在中间）**：

每次分区后，数组大致被分成两半，需要进行 **log N** 层分区。
每层分区总共处理N个元素，所以总步数为 **N × log N**。

<span style="color: yellow">平均情况时间复杂度：O(N log N)</span>

---

**最坏情况（数组已经有序，每次轴都在最边）**：

<span style="color: yellow">[1, 2, 3, 4, 5, 6]</span>

每次分区只减少1个元素，需要进行N层分区。
总步数为：N + (N-1) + (N-2) + ... + 1 = **N²/2**

**最坏情况时间复杂度：O(N²)**

### 复杂度对比图

<img src="./pics/complexity_compare.png" width="700">

- O(N log N)：快速排序平均情况
- O(N²)：冒泡、选择、插入排序
- 当N=100时，O(N log N) ≈ 700，O(N²) = 10000

## 扩展：快速选择 (Quickselect)

快速排序还有一个非常实用的变种——<span style="color: yellow">快速选择</span>。

**问题**：在一个无序数组中，找到第k小的元素。

**朴素方法**：先排序（O(N log N)），再取第k个元素。

**快速选择**：利用分区的思想，平均只需要 <span style="color: yellow">O(N)</span>。

### 快速选择的原理

1. 对数组进行分区，得到轴的位置p
2. 如果p == k，返回arr[p]
3. 如果p > k，在左子数组中找第k小
4. 如果p < k，在右子数组中找第(k-p-1)小

每次分区后，只需要处理<span style="color: yellow">一半的数组</span>，所以平均只需要N + N/2 + N/4 + ... ≈ 2N步。

### 快速选择过程

- 找第2小的元素，分区后轴在第5位
- 只需在左边继续查找
- <span style="color: yellow">不需要排序整个数组。</span>

````python
class QuickSelect:
    """
    快速选择类，用于在无序数组中查找第k小的元素。
    平均时间复杂度：O(N)
    """

    def __init__(self, arr: Sequence[int | float]):
        """
        初始化快速选择器。

        :param arr: 要查询的数组
        """
        self.arr: list[int | float] = list(arr)
        self.length: int = len(arr)

    def partition(self, left_index: int, right_index: int) -> int:
        """
        分区操作（与快速排序相同）。

        :param left_index: 左边界索引
        :param right_index: 右边界索引
        :return: 轴的最终位置
        """
        pivot: int | float = self.arr[right_index]
        left_pointer: int = left_index
        right_pointer: int = right_index - 1

        while True:
            while left_pointer <= right_pointer and self.arr[left_pointer] < pivot:
                left_pointer += 1
            while left_pointer <= right_pointer and self.arr[right_pointer] > pivot:
                right_pointer -= 1
            if left_pointer >= right_pointer:
                break
            self.arr[left_pointer], self.arr[right_pointer] = (
                self.arr[right_pointer],
                self.arr[left_pointer],
            )

        self.arr[left_pointer], self.arr[right_index] = (
            self.arr[right_index],
            self.arr[left_pointer],
        )
        return left_pointer

    def select(
        self, k: int, left_index: int | None = None, right_index: int | None = None
    ) -> int | float:
        """
        查找第k小的元素（k从0开始计数）。

        :param k: 要找第k小的元素（0表示最小，length-1表示最大）
        :param left_index: 搜索范围的左边界
        :param right_index: 搜索范围的右边界
        :return: 第k小的元素的值
        """
        if left_index is None:
            left_index = 0
        if right_index is None:
            right_index = self.length - 1

        # 基准情形：只剩一个元素
        if left_index == right_index:
            return self.arr[left_index]

        # 分区
        pivot_position: int = self.partition(left_index, right_index)

        if k < pivot_position:
            # 第k小在左子数组
            return self.select(k, left_index, pivot_position - 1)
        elif k > pivot_position:
            # 第k小在右子数组(此处为总数组的绝对索引)
            return self.select(k, pivot_position + 1, right_index)
        else:
            # 找到了！
            return self.arr[pivot_position]
````

````python
# 测试快速选择
test_arr3: list[int] = [64, 34, 25, 12, 22, 11, 90]
print(f"数组: {test_arr3}")
print(f"排序后参考: {sorted(test_arr3)}")

# 找第0小（最小）
qs3: QuickSelect = QuickSelect(test_arr3.copy())
print(f"\n第0小（最小）: {qs3.select(0)}")

# 找第3小
qs3 = QuickSelect(test_arr3.copy())
print(f"第3小: {qs3.select(3)}")

# 找第6小（最大）
qs3 = QuickSelect(test_arr3.copy())
print(f"第6小（最大）: {qs3.select(6)}")

# 找中位数
qs3 = QuickSelect(test_arr3.copy())
median_index: int = len(test_arr3) // 2
print(f"中位数（第{median_index}小）: {qs3.select(median_index)}")
````

## 总结

- <span style="color: yellow">分区</span>是快速排序的核心
- 递归解决左右子数组
- 平均O(N log N)，最坏O(N²)
- <span style="color: yellow">快速选择</span>可在O(N)找到第k小元素
