```python
from typing import Any
from collections.abc import Sequence


class OrderedArrayInsert:
    """
    有序数组的插入
    """

    def __init__(self, iterable: Sequence[Any]):
        """
        :param iterable: 有序数组初始化
        """
        self.iterable = list(iterable)
        self.len = len(self.iterable)

    def insert(self, element: Any) -> None:
        """
        插入元素
        :param element: 待插入的元素
        """

        insert_index: int = 0

        for index, value in enumerate(self.iterable):
            if element < value:
                insert_index = index
                break

        self.iterable.append("")
        for i in range(self.len, insert_index - 1, -1):
            self.iterable[i] = self.iterable[i-1]

        self.iterable[insert_index] = element
```

```python
base_array: list[int] = [1, 4, 21, 59]

ordered_array_insert = OrderedArrayInsert(base_array)
ordered_array_insert.insert(19)
```

### 有序数组的查找

回顾一下<span style="color: yellow">常规数组</span>的查找方案


```python
class ArraySearch:
    """
    常规数组的查找
    """

    def __init__(self, iterable: Sequence[Any]):
        """
        :param iterable: 数组初始化
        """
        self.iterable = list(iterable)

    def search(self, element: Any) -> int:
        """
        查找元素
        :param element: 待查找的元素
        :return: 元素的索引，查不到返回-1
        """
        for index, value in enumerate(self.iterable):
            if element == value:
                return index
        return -1
```

```python
array_search = ArraySearch([1, 4, 21, 59])
array_search.search(21)
```

## 变成有序数组，能加速吗？

```python
class OrderedArraySearch:
    """
    有序数组查询
    """
    def __init__(self, iterable: Sequence[Any]):
        """
        :param iterable: 有序数组初始化
        """
        self.iterable = list(iterable)

    def search(self, element: Any) -> int:
        """
        查询元素
        :param element: 待查询的元素
        :return: 元素的索引，查不到返回-1
        """

        for index, value in enumerate(self.iterable):
            if element == value:
                return index
            elif element < value: # <-- 在这里能够加速
                return -1
        return -1
```
