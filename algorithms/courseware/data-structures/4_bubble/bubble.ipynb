### 冒泡排序规则

- 从数组的第一个元素开始，<span style="color: yellow">相邻的两个元素</span>进行比较。
- 如果前一个元素<span style="color: yellow">大于</span>后一个元素，则交换这两个元素的位置。
- 每次比较完相邻元素后，继续向后移动一对，继续比较和交换，直到比较到最后一个元素为止，记为一轮。
- 重复以上过程，但每次比较的范围减少最后一个已排序的位置（因为末尾的元素已经是有序的）。
- 当一轮比较中没有发生任何交换时，说明数组已经有序，则结束。

### 接下来用代码实现一下

```python
class BubbleSort:
    def __init__(self, arr: list[int|float]):
        """
        :param arr: 初始化数组，可以是整数，也可以是浮点数，只要能比较大小就行
        """
        self.arr = arr
        self.len = len(arr)

    def sort(self) -> list[int|float]:
        """
        冒泡排序实现

        :return: 返回已排序的数组
        """

        # 最多只需要执行 self.len - 1 轮
        # 主循环
        for i in range(self.len - 1):

            # 在内部循环外层，记录是否发生过交换
            # 每执行一次内部循环之前，都将是否发生过交换置为 False
            # 如果在内部循环内改变了这个值，则说明在当前轮中发生过交换，则还需要继续执行下一轮
            # 如果在当前轮中未发生过交换，则可以终止
            is_swapped: bool = False

            # 内部循环
            for j in range(self.len - i - 1):
                if self.arr[j] > self.arr[j + 1]:
                    self.arr[j], self.arr[j + 1] = self.arr[j + 1], self.arr[j]
                    # 如果发生过交换，则将 is_swapped 置为 True，下一轮还将继续
                    is_swapped = True

            # 如果内部循环结束了，is_swapped 还没有被改变过，就可以提前结束了
            if not is_swapped:
                break

        return self.arr
```

```python
bubble_arr = BubbleSort([4, 2, 5, 1, 3, 6.26, 1.02])
bubble_arr.sort()
```

### 冒泡排序的时间复杂度是什么?

```python
import numpy as np
import matplotlib.pyplot as plt

x_axis = np.arange(1, 101)
y_axis_n = x_axis
y_axis_n2 = x_axis ** 2
y_axis_logn = np.log(x_axis)
plt.plot(x_axis, y_axis_n, label="n")
plt.plot(x_axis, y_axis_n2, label="n^2")
plt.plot(x_axis, y_axis_logn, label="log(n)")
plt.ylim(0, 100)

plt.legend(labelcolor="black")
```
