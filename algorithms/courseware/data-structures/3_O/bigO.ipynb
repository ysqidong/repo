## 大O记法

- 大O记法不关注具体用多少时间
- 只关注用的步数
- <span style="color: yellow">一般描述的是最差情况</span>

### $O(1)$

- 只需要1步
- 读取数组的元素
- 末位插入、删除

### $O(N)$

- 操作步数与数组大小成<span style="color: yellow">正比</span>
- 数组中间位置的插入、删除
- 数组元素查找

```python
import numpy as np
import matplotlib.pyplot as plt

x_axis: np.ndarray = np.arange(1, 101)
y_axis_linear: np.ndarray = np.arange(1, 101)
y_axis_1: np.ndarray = np.ones(100)

plt.plot(x_axis, y_axis_linear, label="linear")
plt.plot(x_axis, y_axis_1, label="constant")

plt.legend(labelcolor="black")
```

### 问题：$O(1)$一定比$O(N)$快吗？

### $O(logN)$

- 数组规模每乘以2，步数增加1
- 效率略差于$O(1)$，远远好于$O(N)$

```python
import numpy as np
import matplotlib.pyplot as plt

x_axis: np.ndarray = np.arange(1, 101)
y_axis_linear: np.ndarray = np.arange(1, 101)
y_axis_1: np.ndarray = np.ones(100)
y_axis_log: np.ndarray = np.log(x_axis)

plt.plot(x_axis, y_axis_linear, label="linear")
plt.plot(x_axis, y_axis_1, label="constant")
plt.plot(x_axis, y_axis_log, label="log")

plt.legend(labelcolor="black")
```
