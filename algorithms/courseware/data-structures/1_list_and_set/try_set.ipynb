集合与列表高度相似

只是增加了一个规则——<span style="color: yellow">集合中的元素不能重复</span>

## 插入

仍然是这个水果的列表，现在是集合了

<img src="./pics/水果.png" width="700" />

现在想要在集合的末尾插入一个元素

如果是列表，直接插入就可以了

但是集合，先需要检查，这个元素在原始集合里是否存在

我们仍然是要插入一个pear

<img src="./pics/插入-set-1.png" width="700" />


插入之前，我们需要对原始集合的每一个元素检查，是否为pear

<img src="./pics/find_index_0.png" width="700" />


<img src="./pics/find_index_1.png" width="700" />

```python
from collections.abc import Sequence
from typing import Any


class SetInsert:
    """
    向集合中插入元素
    插入前需要检查是否已经有这个元素
    """

    def __init__(self, iterable: Sequence[Any]):
        """
        :param iterable: 初始化的集合，实际类型仍为list
        """
        self.iterable = list(iterable)

    def insert_last(self, element: Any) -> bool:
        """
        在末尾插入一个元素

        :param element: 要插入的元素
        :return: 插入是否成功
        """

        for index, item in enumerate(self.iterable):
            if item == element:
                return False
        self.iterable.append(element)
        return True
```

```python
fruits = ['apple', 'banana', 'cherry', 'orange', 'mango']
set_insert = SetInsert(fruits)
set_insert.insert_last('pear')
```

```python
print(set_insert.iterable)
```
