### 我们把这套方法迁移到有序数组的查找

如果想要查找到7

<img src="./pics/二分-1.png" width="900" />

```python
class BinarySearch:
    """
    二分查找
    """
    def __init__(self, ordered_array: list[int]):
        """
        :param ordered_array: 有序数组初始化
        """
        self.ordered_array = ordered_array

        self.len = len(self.ordered_array)

    def search(self, target: int) -> int:
        """
        使用二分查找方法找到目标

        :param target: 查找的目标
        :return: 查找到的索引，没查到返回-1
        """

        left: int = 0
        right: int = self.len - 1

        while left <= right:
            mid: int = (left + right) // 2
            mid_value: int = self.ordered_array[mid]

            if mid_value == target:
                return mid
            elif mid_value < target:
                left = mid + 1
            else:
                right = mid - 1

        return -1
```

```python
ordered_array: list[int] = [1,4,5,7,8,10,12,13,14,16]
binary_search = BinarySearch(ordered_array)
```

```python
binary_search.search(7)
```

### 二分查找的时间复杂度

二分查找每次查找至少排除一半

也就是说——数组长度增长一倍，二分查找只会增加一步

```python
import numpy as np
import matplotlib.pyplot as plt

x_axis: np.ndarray = np.arange(1, 101)
y_axis_linear_search: np.ndarray = x_axis
y_axis_binary_search: np.ndarray = np.log2(x_axis)

plt.plot(x_axis, y_axis_linear_search, label="Linear Search")
plt.plot(x_axis, y_axis_binary_search, label="Binary Search")

legend = plt.legend(labelcolor="black")
```

有序数组的二分查找速度明显快于线性查找

量越大快得越明显

但是插入却会很慢
