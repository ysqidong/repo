## 线性查找

按照索引逐个查找是否为banana

- 如果是，就输出索引，跳出循环
- 如果不是就继续下一个

```python
# 将从列表中查找水果的过程封装为一个函数
def find_fruit(fruit_name: str) -> int:
    """
    找到指定水果，并返回索引，如果没找到则返回-1

    :param fruit_name: 水果名字
    :return: 索引
    """
    index: int
    fruit: str
    for index, fruit in enumerate(fruits):
        if fruit == fruit_name:
            return index
            break # 如果没有这一行，速度会存在差异

    return -1
```

```python
# 查找苹果
find_fruit("apple") # 需要一次循环
```

```python
# 查找芒果
find_fruit("mango") # 需要五次循环
```

```python
# 查找西瓜(找不到)
find_fruit("watermelon") # 需要五次循环
```

## 上面的案例数量太少，体现不出性能差距

如果换一个大列表，我们就能感受到差距了

```python
# 构建一个足够大的列表
number_list: list[int] = [i for i in range(100_000_000)]

def find_number(number: int) -> int:
    """
    找到指定数字，返回索引，未找到返回-1

    :param number: 要找的数字
    :return: 索引
    """
    index: int
    num: int
    for index, num in enumerate(number_list):
        if num == number:
            return index
            break # 如果没有这一行，速度会存在差异

    return -1
```

```python
find_number(1)
```

```python
find_number(99_999_999)
```

```python
find_number(-1)
# 一个如此规整的数字，人类一眼就能看出不在列表中
# 机器任然要全面搜索一遍
```

## 我们把上面所有的查询案例抽象一层

给定一个可迭代对象，以及要查询的元素，返回索引，未找到返回-1

```python
from typing import Iterable, Any

class FindIndex:
    """
    给定一个可迭代对象，以及要查询的元素，返回索引，未找到返回-1
    """
    def __init__(self, iterable: Iterable[Any]):
        """
        :param iterable: 可迭代对象，查找其内部元素
        """
        self.iterable: list[Any] = list(iterable)

    def find(self, element: Any) -> int:
        """
        找到指定元素，返回索引，未找到返回-1

        :param element: 要找的元素
        :return: 索引
        """
        index: int
        element_in_iterable: Any
        for index, element_in_iterable in enumerate(self.iterable):
            if element_in_iterable == element:
                return index
                break

        return -1
```

```python
FindIndex(fruits).find("banana")
```

```python
FindIndex(number_list).find(99_999_999)
```

如果我们想在列表开头插入数据，我们就需要先对整个列表的每一个元素进行一次移动

<span style="color: yellow;">最差的情况</span>，向一个含有N个元素的列表中插入一个元素，需要N+1步

```python
l = []
l.append(1)
```

```python
from typing import Iterable, Any

class InsertItem:
    """
    向列表中插入元素
    """
    def __init__(self, iterable: Iterable[Any]):
        """
        :param iterable: 可迭代对象，向其内部插入元素
        """
        self.iterable = list(iterable)

    def insert(self, element: Any, index_to_insert: int) -> list[Any]:
        """
        向列表中插入元素

        :param element: 要插入的元素
        :param index_to_insert: 插入的索引
        """

        length_of_iterable: int = len(self.iterable)
        # 提前在末尾留一个空位
        self.iterable.append("")
        # 从后往前移动元素
        # 注意，不能颠倒顺序，否则会覆盖原来的元素
        for i in range(length_of_iterable, index_to_insert, -1):
            self.iterable[i] = self.iterable[i - 1]

        self.iterable[index_to_insert] = element

        return self.iterable
```

```python
insert_item = InsertItem(fruits)
insert_item.insert("pear", 2)
```

## 时间复杂度

目的——衡量算法执行时间随输入规模增长而变化的趋势

- 我们通常需要考虑最差的结果
- 有时需要考虑平均结果
- 而不太考虑最好的结果


如果我们有一个长度为N的列表，对它的某个操作的步数会和长度N有关系

这个关系的表达式，我们只取最高项，并忽略常数项或常数因子

例如：

某个操作需要的步数需要$2N^2 + 3N + 4$步

它的时间复杂度应该为$O(N^2)$
