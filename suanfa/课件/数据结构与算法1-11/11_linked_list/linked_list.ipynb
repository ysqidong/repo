## 什么是链表？

链表是一种<span style="color: yellow">基于结点的数据结构</span>。与数组不同，链表的元素在内存中<span style="color: yellow">不需要连续</span>。

### 链表的基本组成

1. **结点 (Node)**：存储数据的基本单元
2. **链 (Link)**：指向下一结点的内存地址
3. **头指针 (Head)**：指向链表第一个结点的引用

### 链表的存储方式

- **数组**：元素在内存中连续存储
- **链表**：结点分散在内存各处，通过链连接

<span style="color: yellow">关键特性</span>：每个结点保存数据和指向下一结点的地址。

### 内存结构对比

**数组的内存布局**（连续存储）：

```
内存地址:  1000   1001   1002   1003   1004
数组:      "a"    "b"    "c"    "d"    "e"  
           ↑
         数组起始地址
```

**链表的内存布局**（分散存储，通过链连接）：

```
结点1              结点2              结点3              结点4
┌──────┬─────┐    ┌──────┬─────┐    ┌──────┬─────┐    ┌──────┬─────┐
│ "a"  │  ───┼───►│ "b"  │  ───┼───►│ "c"  │  ───┼───►│ "d"  │null │
└──────┴─────┘    └──────┴─────┘    └──────┴─────┘    └──────┴─────┘
  ↑                                                      ↑
 1901                                                   5237
(内存地址分散)                                          (分散各处)
```

- 数组：知道起始地址和索引，一步计算出元素位置
- 链表：必须从第一个结点开始，顺着链逐个遍历

````python
class Node:
    """
    链表结点类。
    
    每个结点包含：
    - data: 存储的数据
    - next_node: 指向下一个结点的引用（链）
    """

    def __init__(self, data: str | int | float):
        """
        初始化结点。

        :param data: 结点存储的数据
        """
        self.data: str | int | float = data
        self.next_node: Node | None = None

    def __repr__(self) -> str:
        """返回结点的字符串表示。"""
        return f"Node({self.data!r})"


class LinkedList:
    """
    链表类。
    
    维护一个指向第一个结点的头指针，
    通过头指针可以遍历整个链表。
    """

    def __init__(self):
        """
        初始化空链表。
        """
        self.first_node: Node | None = None

    def __repr__(self) -> str:
        """返回链表的字符串表示。"""
        if not self.first_node:
            return "LinkedList([])"
        
        # 初始化空列表
        nodes: list[str] = []
        # 指定链表的查询起始点
        # 可以不在类初始化时完成
        current: Node | None = self.first_node
        # 从第一个Node开始
        while current:
            nodes.append(str(current.data))
            # 如果有下一个节点，while循环继续
            current = current.next_node
        return f"LinkedList([{', '.join(nodes)}])"
````

````python
# 创建链表: "once" -> "upon" -> "a" -> "time"

# 创建结点
node_1 = Node("once")
node_2 = Node("upon")
node_3 = Node("a")
node_4 = Node("time")

# 连接结点
# 注意：Node.next_node应该是一个Node，而不是str
node_1.next_node = node_2
node_2.next_node = node_3
node_3.next_node = node_4

# 创建链表，指定头结点
list_ = LinkedList()
list_.first_node = node_1

print(f"链表: {list_}")
print("\n遍历链表:")
current = list_.first_node
while current:
    print(f"  {current.data}")
    current = current.next_node
````

## 操作1：读取 (Read)

### 数组 vs 链表

- **数组**：通过索引直接访问，O(1)
- **链表**：必须从第一个结点开始遍历，O(N)

### 为什么链表读取是O(N)？

链表结点在内存中分散存储，程序只知道第一个结点的地址。
要读取索引2的元素，需要：

```
步骤1: 从first_node开始（索引0）
步骤2: 通过链找到下一个结点（索引1）
步骤3: 再通过链找到下一个结点（索引2）
```

<span style="color: yellow">最坏情况</span>：读取最后一个元素，需要遍历N个结点 → **O(N)**

````python
def read(self, index: int) -> str | int | float | None:
    """
    读取指定索引位置的元素。
    
    时间复杂度: O(N)
    
    :param index: 要读取的索引（从0开始）
    :return: 索引位置的元素值，索引越界返回None
    """
    current_node: Node | None = self.first_node
    current_index: int = 0
    
    # 有了索引，仍然需要遍历才能到达索引位置
    while current_index < index:
        if not current_node:
            return None  # 索引越界，无法在遍历之前就发现越界，没有分配整段内存
        current_node = current_node.next_node
        current_index += 1
    
    # current_index += 1在遍历最后一次结束前执行过
    return current_node.data if current_node else None

# 绑定方法到类
LinkedList.read = read

# 测试读取
print(f"链表: {list_}")
print(f"\n读取索引0: {list_.read(0)}")   # "once"
print(f"读取索引1: {list_.read(1)}")   # "upon"
print(f"读取索引2: {list_.read(2)}")   # "a"
print(f"读取索引3: {list_.read(3)}")   # "time"
print(f"读取索引10: {list_.read(10)}")  # None（越界）
````

## 操作2：查找 (Search)

查找：找出某个值在链表中的索引。

### 链表 vs 数组

两者查找效率**相同**，都是 O(N)。

因为无论链表还是数组，都需要从第一个元素开始逐个比较：

```
从头开始 → 比较 → 不匹配 → 下一个 → 比较 → 不匹配 → 下一个...
```

<span style="color: yellow">最坏情况</span>：查找的元素在最后或不存在 → **O(N)**

````python
def index_of(self, value: str | int | float) -> int | None:
    """
    查找指定值的索引。
    
    时间复杂度: O(N)
    
    :param value: 要查找的值
    :return: 值的索引，未找到返回None
    """
    current_node: Node | None = self.first_node
    current_index: int = 0
    
    # 如果没有对应元素，则从头查到尾
    while current_node:
        if current_node.data == value:
            return current_index
        current_node = current_node.next_node
        current_index += 1
    
    return None  # 未找到

# 绑定方法
LinkedList.index_of = index_of

# 测试查找
print(f"链表: {list_}")
print(f"\n查找 'a': 索引 {list_.index_of('a')}")
print(f"查找 'time': 索引 {list_.index_of('time')}")
print(f"查找 'hello': {list_.index_of('hello')}")  # None
````

## 操作3：插入 (Insert)

链表插入的效率与位置有关：

### 在链表开头插入 - O(1)

```
原链表:  ["b"] → ["c"] → ["d"]

插入 "a" 到开头:
1. 创建新结点 ["a"]
2. 让 ["a"] 的 next 指向 ["b"]
3. 更新 first_node 为 ["a"]

结果:   ["a"] → ["b"] → ["c"] → ["d"]
```

**只需修改指针，无需移动其他元素** → <span style="color: yellow">O(1)</span>

### 在链表中间/末尾插入 - O(N)

需要先遍历找到插入位置的前一个结点，再修改指针。

<span style="color: yellow">原因</span>：需要先找到插入位置 → 遍历N个结点 → **O(N)**

````python
def insert_at_index(self, index: int, value: str | int | float) -> bool:
    """
    在指定索引位置插入新元素。
    
    时间复杂度:
    - 索引0（开头）: O(1)
    - 其他位置: O(N)
    
    :param index: 插入位置索引
    :param value: 要插入的值
    :return: 插入成功返回True，索引越界返回False
    """
    new_node = Node(value)
    
    # 在开头插入
    if index == 0:
        new_node.next_node = self.first_node
        self.first_node = new_node
        return True
    
    # 准备好查询节点
    current_node: Node | None = self.first_node
    current_index: int = 0
    
    # 找到插入位置的前一个结点
    # ["hello"] -->  ["world"] --> ["！"]
    #   0               1            2
    # 每一轮while循环向右移动一个node
    # while循环结束的时候正好停在index - 1
    while current_index < index - 1:
        if not current_node:
            return False  # 索引越界
        current_node = current_node.next_node
        current_index += 1
    
    # 循环结束后补充检查一次索引越界
    if not current_node:
        return False
    
    # 插入新结点
    # 就像插队一样
    new_node.next_node = current_node.next_node
    current_node.next_node = new_node
    return True

# 绑定方法
LinkedList.insert_at_index = insert_at_index

# 测试插入
test_list = LinkedList()
test_list.insert_at_index(0, "world")  # 开头插入
test_list.insert_at_index(0, "hello")  # 开头插入
test_list.insert_at_index(2, "!")     # 末尾插入
test_list.insert_at_index(2, " beautiful")  # 中间插入

print(f"插入后: {test_list}")
````

## 操作4：删除 (Delete)

链表删除的效率同样与位置有关：

### 在链表开头删除 - O(1)

```
原链表:  ["a"] → ["b"] → ["c"] → ["d"]

删除第一个元素:
1. 保存 first_node 的数据（"a"）
2. 将 first_node 更新为第二个结点

结果:   ["b"] → ["c"] → ["d"]
```

**只需修改头指针** → <span style="color: yellow">O(1)</span>

对比数组：删除第一个元素需要将所有后续元素左移 → O(N)

### 在链表中间/末尾删除 - O(N)

需要先遍历找到要删除结点的前一个结点。

```
删除索引2（"c"）:
1. 找到索引1的结点（"b"）
2. 让 ["b"] 的 next 指向 ["d"]
3. ["c"] 不再被引用
```

## 链表 vs 数组 性能对比

| 操作 | 数组 | 链表 | 备注 |
|------|------|------|------|
| **读取** | O(1) | O(N) | 数组直接访问，链表需遍历 |
| **查找** | O(N) | O(N) | 两者都需要逐个比较 |
| **前端插入** | O(N) | O(1) | 链表只需改指针 |
| **前端删除** | O(N) | O(1) | 链表只需改头指针 |
| **末尾插入** | O(1) | O(N) | 数组 amortized O(1) |
| **末尾删除** | O(1) | O(N) | 链表需先找到倒数第二个 |
| **中间插入** | O(N) | O(N) | 链表需先找到位置 |
| **中间删除** | O(N) | O(N) | 链表需先找到位置 |


- <span style="color: yellow">数组优势</span>：随机访问快 O(1)
- <span style="color: yellow">链表优势</span>：头部操作快 O(1)，遍历过程中删除快O(N)

## 链表实战：批量删除无效邮件

**场景**：整理电子邮件列表，删除格式无效的地址。

### 算法流程

1. 遍历邮件列表
2. 对每个地址验证格式（用正则表达式）
3. 如果无效，从列表中删除

### 性能对比（1000个地址，100个无效）

**使用数组**：
- 遍历检查：1000步
- 删除操作：每次删除需左移元素 → 平均500步 × 100次 = 50,000步
- **总计：约51,000步**

**使用链表**：
- 遍历检查：1000步
- 删除操作：只需改指针 → 1步 × 100次 = 100步
- **总计：约1,100步**

<span style="color: yellow">结论</span>：链表在需要频繁删除的场景下优势明显！

## 双向链表 (Doubly Linked List)

链表的升级版，每个结点有两个链：

```
┌─────────┬───────┬─────────┐
│ previous│  data │  next   │
└─────────┴───────┴─────────┘
```

### 优势

- 可以双向遍历（从头到尾，或从尾到头）
- 直接访问最后一个结点
- **末尾插入和删除都是 O(1)**

### 适用场景

<span style="color: yellow">队列的最佳底层实现</span>：
- enqueue（末尾插入）：O(1)
- dequeue（开头删除）：O(1)

````python
class DNode:
    """双向链表结点"""
    def __init__(self, data: str | int | float):
        self.data = data
        self.prev: DNode | None = None
        self.next: DNode | None = None

class DoublyLinkedList:
    """双向链表"""
    def __init__(self):
        self.first: DNode | None = None
        self.last: DNode | None = None
    
    def insert_at_end(self, value: str | int | float) -> None:
        """在末尾插入 - O(1)"""
        new_node = DNode(value)
        if not self.first:  # 空链表
            self.first = self.last = new_node
        else:
            # 新节点的prev指向原来的队尾
            new_node.prev = self.last
            # 原队尾的next指向新节点
            self.last.next = new_node
            # 新队尾指向新节点
            self.last = new_node
    
    def remove_from_front(self) -> str | int | float | None:
        """从开头删除 - O(1)"""
        if not self.first:
            return None
        data = self.first.data
        self.first = self.first.next
        if self.first:
            self.first.prev = None
        else:
            self.last = None
        return data

# 用双向链表实现队列
class Queue:
    """基于双向链表的队列"""
    def __init__(self):
        self._dll = DoublyLinkedList()
    
    def enqueue(self, value: str | int | float) -> None:
        """入队 - O(1)"""
        self._dll.insert_at_end(value)
    
    def dequeue(self) -> str | int | float | None:
        """出队 - O(1)"""
        return self._dll.remove_from_front()
    
    def is_empty(self) -> bool:
        return self._dll.first is None

# 测试队列
queue = Queue()
for i in range(1, 4):
    queue.enqueue(i)
    print(f"入队: {i}")

print("\n开始出队:")
while not queue.is_empty():
    print(f"  出队: {queue.dequeue()}")
````

### 何时使用链表？

1. **频繁在头部插入/删除**
2. **需要遍历并删除多个元素**
3. **不需要随机访问**
4. **数据量动态变化大**

## 实际场景中，链表使用多不多
1. <span style="color: yellow">缓存局部性差</span>
  - 链表：节点分散在堆内存各处，CPU访问下一个节点时几乎一定会发生缓存未命中，需要从主存重新加载数据。
  - 动态数组：元素在内存中连续存放，CPU可以利用空间局部性，一次性将多个元素加载到缓存行中。
  - 在现代CPU架构下，一次主存访问的延迟相当于几百个时钟周期。即使链表在算法复杂度上看起来不错，实际运行时间往往比数组慢数倍甚至数十倍。

  2. <span style="color: yellow">内存开销大</span>
  - 每个链表节点除了存储数据本身，还需要额外存储一个或两个指针。如果存储的是小块数据（如int），指针开销可能超过数据本身，导致内存利用率极低。

  3. <span style="color: yellow">不支持随机访问</span>
  - 按索引查找：$O(n)$
  - 排序：效率远低于数组
  - 二分查找：无法直接使用

  4. <span style="color: yellow">内存分配器开销和碎片</span>
  - 链表的插入删除往往伴随着频繁的 malloc / free
  - 动态数组虽然也可能触发扩容重分配，但分配频率远低于链表。

  5. <span style="color: yellow">对CPU分支预测和预取不友好</span>
  - 现代CPU擅长顺序访问和分支预测，链表的跳转式访问会破坏指令流水线和数据预取
