## 插入排序

```python
class InsertionSort:
    def __init__(self, arr: list[int|float]):
        """
        :param arr: 初始化数组
        """
        self.arr = arr
        self.len = len(arr)

    def sort(self) -> list[int | float]:
        """
        插入排序实现

        :return: 返回排序后数组
        """

        # 从第二位（索引为1）开始
        for index in range(1, self.len):

            # 从每一次的索引位置开始
            position = index
            # 保存当前索引位置的值
            temp_value = self.arr[index]
            # 循环判断，位置左侧的值是否大于此轮次的索引位置的值
            # 如果大于，就把左侧的值赋值给索引位置，并把索引位置减1
            # 反之，此轮结束，索引位置依照for循环向下一个
            while (position > 0) and (self.arr[position -1] > temp_value):
                self.arr[position]=self.arr[position -1]
                position=position -1

            self.arr[position]=temp_value

        return self.arr
```

```python
insertion_sort = InsertionSort([7, 2, 3, 4, 1])
insertion_sort.sort()
```

### 插入排序的时间复杂度是什么？

1 + 2 + 3 + ... + (n-1) = n(n-1)/2 ≈ O(n²)
